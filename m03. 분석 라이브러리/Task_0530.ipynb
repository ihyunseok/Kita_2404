#### Task1_0530. 고객 정보 데이터프레임과 구매 내역 데이터프레임을 결합하여 각 고객이름(name) 별 총 지출을 계산하여 출력하세요.

```python
import pandas as pd

# 고객 정보 데이터프레임 생성
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'David']
})

# 구매 내역 데이터프레임 생성
purchases = pd.DataFrame({
    'purchase_id': [101, 102, 103, 104, 105],
    'customer_id': [1, 2, 1, 3, 2],
    'amount': [250, 150, 200, 300, 350]
})
print(customers,'\n')
print(purchases)
```

```python
c = customers.join(purchases,lsuffix = '_1', rsuffix = '_2', how = 'inner')
c
```

```python
def amount_sum(x):
    return x.sum()
```

```python
c.columns
```

```python
c['customer_amount'] = c.groupby('name')['amount'].transform(amount_sum)
c
```

#### Task8_0530. 승객의 가족 크기(동반한 형제자매/배우자 수와 부모/자녀 수의 합)가 생존에 영향을 미치는지를 탐색하여 결과를 출력하세요.

```python
# 데이터프레임 읽기
file_path = 'titanic3.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.columns
```

```python
con1 = pd.concat([df['survived'], df['parch']], axis = 1)
con1
```

```python
df['survived'].count()
```

```python
def aliveordeath(x):
      # 부모/자녀 수의 합
      if (x['survived'] == 1) and (x['parch'] >= 2):
           return '부모/자녀와 함께 생존'

      # 형제자매/배우자 수
      if (x['survived'] == 1) and (x['parch'] >= 1):
           return '형제자매/배우자와 함께 생존'

      # 혼자 탄 사람
      if (x['survived'] == 1) and (x['parch'] == 0):
           return '혼자 생존'

      else:
          return '사망'

con1['tally'] = con1.apply(aliveordeath, axis = 1)
con1.head(20)
```

```python
def alive_counts(x):
    if x['tally'] =='부모/자녀와 함께 생존':
        return 3
    elif x['tally'] == '형제자매/배우자와 함께 생존' :
        return 2
    elif x['tally'] == '혼자 생존':
        return 1
    else :
        return 0

con1['tally_count'] = con1.apply(alive_counts, axis = 1)
con1.head(20)
```

```python
# 승객의 가족 크기(동반한 형제자매/배우자 수와 부모/자녀 수의 합)가
# 생존에 영향을 미치는지를 탐색하여 결과를 출력하세요.
# 동반한 형제자매/배우자 수와 부모/자녀 수의 합 / 전체 생존자 수 * 100 = 생존자 중 가족있는 비율
anyone = df['survived'].count()
print(f"전체 인원 수 : {anyone}명")

count_live_anyone = con1['survived'].sum()
print(f"전체 생존자의 수 : {count_live_anyone}명 \n")

count_live_alone_count = con1[con1['tally_count'] == 1].shape[0]
print(f"동행이 없는 생존자의 수 : {count_live_alone_count}명 \n")

count_live_two_count = con1[con1['tally_count'] == 2].shape[0]
print(f"동행이 1명 있는 생존자의 수 : {count_live_two_count}명 \n")

count_live_family_count = con1[con1['tally_count'] == 3].shape[0]
print(f"동행이 2명 이상 있는 생존자의 수 : {count_live_family_count}명 \n")

p0 = round(((count_live_two_count + count_live_family_count) / anyone) * 100, 2)
p1 = round(((count_live_two_count + count_live_family_count) / count_live_anyone) * 100, 2)
p2 = round((count_live_alone_count / count_live_anyone) * 100, 2)
p3 = round((count_live_alone_count / anyone) * 100, 2)
p4 = round((count_live_anyone / anyone) * 100, 2)

print(f"생존자중, 동행이 있었던 생존자 비율 : {p1}%")
print(f"생존자중, 동행이 없었던 생존자 비율 : {p2}%")
print(f"전체 탑승 인원 대비 동행이 있었던 생존자 비율 : {p0}%")
print(f"전체 탑승 인원 대비 동행이 없었던 생존자 비율 : {p3}%")
print(f"전체 탑승 인원 대비 생존율 : {p4}%")
```
