Task1_0527. 4개의 Series를 결합하여 데이터프레임을 생성하여 출력하세요.
- 출력한 데이터프레임에서 추출하고 싶은 5개의 데이터를 인덱싱하여 출력
- null값을 3개 추가
- null값의 개수를 확인하고 삭제
- 2개의 데이터를 수정
- 1개의 행을 삭제

```python
id = np.arange(1, 11)
i2 = pd.Series(id)

gender = np.random.randint(2, size = 101)
g2 = pd.Series(gender)

age = np.random.randint(1,11, size = 101)
a2 = pd.Series(age)

region = np.random.randint(1,11, size = 101)
r2 = pd.Series(region)

df = pd.concat([i2, g2, a2, r2], axis = 1)
print(df.head())

#  출력한 데이터프레임에서 추출하고 싶은 5개의 데이터를 인덱싱하여 출력
print(df.loc[0, 2])
print(df.loc[1, 3])
print(df.loc[4, 0])
print(df.loc[2, 0])
print(df.loc[3, 1])
df.loc[:, 0] = 10
print(df)

# null값을 3개 추가
df.loc[3,0] = np.nan
df.loc[3,2] = np.nan
df.loc[7,3] = np.nan
print(df.head(20))
print()


# null값의 개수를 확인하고 삭제
df_num = df.isnull().sum()
print(df_num)

df_drop = df.dropna()
print(df_drop)

# 2개의 데이터를 수정
df_drop.loc[0, 1] = '수정'
df_drop.iloc[0, 2] = '수정'
print(df_drop)

# 1개의 행을 삭제
df_drop.drop([0], axis = 0, inplace = True)
print(df_drop)
```

Task2_0527. 학생별 성적 데이터셋으로 의미있게 데이터 셋을 수정하세요.(아래 사항 반영)
- S1 ~ S10은 평균 점수를 기준으로 1등급에서 10등급이고 등급간 점수 차는 5점
- 결시자가 국어 3명, 수학 2명 있음
- 영어, 수학의 평균 점수가 국어 대비 5점 낮음

```python
import pandas as pd
import numpy as np

np.random.seed(0)
df = np.random.randint(1,11, size=(10, 5))
df=pd.DataFrame(df,index=['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10'],
               columns=['국어','영어','수학','과학','사회'])

df.loc['s2', '국어'] = np.nan
df.loc['s3', '국어'] = np.nan
df.loc['s9', '국어'] = np.nan
df.loc['s3', '수학'] = np.nan
df.loc['s5', '수학'] = np.nan

df1 = df * 5

df1.mean()
df1
```

```python
# 등급간 점수 차는 5점
# df.iloc[0] # s1
# df.iloc[1] # s2

for i in range(10):
    df.iloc[i] = 90 - i * 5
df
```

```python
# 영어, 수학의 평균 점수가 국어 대비 5점 낮음
# df.iloc[0, 1:3] → 영어 85 / 수학 85
# df.iloc[0, 0] - 5 → 국어 90
for i in range(10):
    df.iloc[i, 1:3] = df.iloc[i, 0] - 5
df
```

```python
# 결시자가 국어 3명, 수학 2명 있음
a = df.copy()
a.iloc[7:, 0] = np.nan
a.iloc[8:, 2] = np.nan
a
```

Task3_0527. df1에서 결측값 처리 후 딕셔너리, 리스트로 변환하여 출력하세요.

```python
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Math": [85, np.nan, 88, 90, 76],
    "English": [92, 85, 89, np.nan, 80],
    "Science": [78, 90, 95, 85, 89],
    "History": [88, 92, 85, 91, np.nan]
})
# 결측값 처리
# df1 = df1.fillna(0)
# df1

# 딕셔너리 및 리스트로 변환 시
# 먼저 해당 데이터프레임을 1-1. 배열로 바꾼 후 :
# df_value = df.values or df_arrays = df.to_numpy()
# 2. 리스트로 바꾼 후 : df_list = df_value.tolist()
# 1 - 2. 딕셔너리로 바꿀 수 있다. df_dict = df.to_dict('list')
df1_value = df1.values
print(df1_value, '\n')

df1_list = df1_value.tolist()
print(df1_list, '\n')

df1_dict = df1.to_dict('list')
print(df1_dict,'\n')
```

```python
# Columns
df1.columns
```

```python
# 숫자 열에 대해 결측값을 각 열의 평균값으로 대체
df2 = df1.copy()
columns = ['Math', 'English', 'Science', 'History']
for column in columns:
    df2[column].fillna(df2[column].mean(), inplace = True)
df2
```

```python
df2_value = df2.values
print(df2_value, '\n')

df2_list = df2_value.tolist()
print(df2_list, '\n')

df2_dict = df2.to_dict('list')
print(df2_dict,'\n')
```

```python
# 배열로 변환
array1 = df2.to_numpy()
print(array1, '\n')

# 딕셔너리로 변환
dict1 = df2.to_dict(orient = 'list')
print(dict1, '\n')

# 리스트로 변환
list1 = df2.values.tolist()
print(list1, '\n')
```

Task4_0527. df2에서 아래와 같이 데이터 필터링 및 정렬 후 배열, 딕셔너리, 리스트로 변환하여 출력하세요.
- Math 점수가 80 이상인 학생만 선택
- English 점수를 기준으로 내림차순 정렬

```python
df2 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Math": [85, 79, 88, 90, 76],
    "English": [92, 85, 89, 93, 80],
    "Science": [78, 90, 95, 85, 89],
    "History": [88, 92, 85, 91, 84]
})
df2_math = df2.loc[df2.Math >= 80,['Name', 'Math', 'English', 'Science', 'History']]

df2_math_englishsort = df2_math.sort_values(['English'])

# 데이터프레임(DF) → 배열(Value) 로 변경
df2_value = df2_math_englishsort.values
print(df2_value, '\n')

# 데이터프레임(DF) → 리스트(list) 로 변경
df2_list = df2_value.tolist()
print(df2_list, '\n')

# 데이터프레임(DF) → 딕셔너리(dict) 로 변경
df2_dict = df2_math_englishsort.to_dict('list')
print(df2_dict)

df2_math_englishsort
```
