#### Task01_0513. 클래스 구성 요소와 관련 아래 사항에 대한 사례를 작성하세요.
- 클래스 선언
- 생성자
- 속성(attributes)
- 메서드
- 상속
- 인스턴스화(객체 만들기)
- 클래스 변수
- 매직 메서드(특수 메서드)

```python
class Remain_food:
    def __init__(self, name, amount):
        self.name = name
        self.amount = amount

    def in_treck(self):
        if self.amount > 100:
            return "음식이 맛없었나요?"
        elif self.amount > 50:
            return "음식이 애매했나요?."
        else:
            return "음식이 맛있었나요?"

class Western(Remain_food):
    def Call(self):
        return f"남긴 음식 / 양식 {self.name} : {self.amount}kg"

class Chinese(Remain_food):
    def Call(self):
        return f"남긴 음식 / 중식 {self.name} : {self.amount}kg"

class KoreanFood(Remain_food):
    def Call(self):
        return f"남긴 음식 / 한식 {self.name} : {self.amount}kg"

Steak = Western('스테이크', 110)
JJamBBong = Chinese('짬뽕', 60)
GimBob = KoreanFood('김밥', 10)

print(f"{Steak.in_treck()}\n{Steak.Call()}")
print()
print(f"{JJamBBong.in_treck()}\n{JJamBBong.Call()}")
print()
print(f"{GimBob.in_treck()}\n{GimBob.Call()}")
```

#### Task2_0513. 두개의 수를 입력한 후 두개의 수에 대한 사칙연산을 수행하세요.(0을 입력한 경우 다시 입력하도록 조치)


```python
class Math:
    def __init__(self):
        self.num1 = float(input("첫 번째 숫자를 입력하세요. : "))
        self.num2 = float(input("두 번째 숫자를 입력하세요. : "))

    def add(self):
        if self.num1 == 0 or self.num2 == 0:
            return "숫자를 다시 입력해주세요."
        else:
            num = self.num1 + self.num2
            return num

    def sub(self):
        if self.num1 == 0 or self.num2 == 0:
            return "숫자를 다시 입력해주세요."
        elif self.num1 > self.num2:
            num = self.num1 - self.num2
            return num
        else :
            num = self.num2 - self.num1
            return num

    def mul(self):
        if self.num1 == 0 or self.num2 == 0:
            return "숫자를 다시 입력해주세요."
        else:
            num = self.num1 * self.num2
            return num

    def div(self):
        if self.num1 == 0 or self.num2 == 0:
            return "숫자를 다시 입력해주세요."
        else:
            num = self.num1 / self.num2
            return num

add = Math()
sub = Math()
mul = Math()
div = Math()

print(f" - Add : {add.add()}")
print(f" - Sub : {sub.sub()}")
print(f" - Mul : {mul.mul()}")
print(f" - Div : {div.div()}")
```

#### Task3_0513. 파이썬 오버라이딩, 오버로딩, 다형성의 사례를 만들어 보세요.


```python
class Country:
    def sayhello(self):
        return "미치셨습니까? 휴먼?"

class Korean(Country):
    def sayhello(self, mood = 'Sad'):
        if mood == 'Sad':
            return "안녕.."
        else:
            return "안녕하세요~"

class Chinese(Country):
    def sayhello(self):
        return "니하오"

class USA(Country):
    def sayhello(self):
        return "하이"

def human_hello(country):
    print(country.sayhello())

korean = Korean()
chinese = Chinese()
usa = USA()3

print(korean.sayhello('Sad'))
human_hello(chinese)
human_hello(usa)
```
