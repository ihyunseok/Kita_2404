## 제어문(control statement)은 프로그램의 흐름을 제어하는데 사용

#### 조건문 (Conditional Statements):
- if: 특정 조건이 참(True)일 때 코드 블록을 실행
    - 조건은 비교 연산자(==, !=, <, >, <=, >=)와 논리 연산자(and, or, not)를 사용하여 표현할 수 있다
- elif: 이전의 if 또는 elif 조건이 거짓(False)일 때 다른 조건을 검사
    - elif는 'else if'의 축약형으로, 이전 if 또는 elif 문의 조건이 거짓(False)일 때 다른 조건을 검사
    - 여러 elif 문을 사용하여 다양한 조건을 검사할 수 있다
- else:
    - 이전의 if 또는 elif 문의 조건이 모두 거짓일 때 실행할 코드 블록을 지정

```python
x = 10
if x > 5:
  print('x는 5보다 큽니다.')
elif x == 5:
  print('x는 5입니다.')
else:
  print('x는 5보다 작습니다.')
```

```python
x = 10
if x >= 5:
  print('x는 5보다 크거나 같습니다.')
else:
  print('x는 5보다 작습니다.')
```

```python
# 각 조건이 서로 독립적일 때 if문을 연속해서 사용
x = 10; y = 5

if x > 5:
  print('x가 5보다 큽니다.')
if y < 10:
  print('y는 10보다 작습니다.')
if x > y:
  print('x는 y보다 큽니다.')
```

```python
x = 5; y = 2

if x > 5:
  if y >= 3:
    print('x가 5보다 크고, y는 3보다 크거나 같습니다.')
  else:
    print('x는 5보다 크고, y는 3보다 작습니다.')
else:
    print('x는 5보다 작거나 같습니다.')
```

```python
def vending_machine(money):
  don = money // 700
  for i in range(0, don+1) :
    print(f'음료수 {i}개, 잔돈 = {3000-(700*i)}원')
vending_machine(3000)
```

```python
# Q. "수입(income)과 지출금액(spend)을 비교해서 수입이 많으면 저축 증가, 반대면 빚이 증가, 같으면 현상 유지" 프로그램 작성

income = int(input('당신의 수입을 작성하세요. : '))
spend = int(input('당신의 매달 지출 금액을 작성하세요. : '))

if income > spend:
  message = "저축을 더 많이 해주세요!"
elif income < spend:
  message = "빚이 증가하고 있어요! 지출을 줄이세요!"
else:
  message = "현상을 유지하고 있어요! 지출을 조금만 줄여보세요!"
print(message)
```

#### 반복문 (Loop Statements):
- for: 시퀀스(리스트, 튜플, 문자열 등)의 아이템을 순회하면서 코드 블록을 실행
- while: 특정 조건이 참인 동안 코드 블록을 반복하여 실행

```python
for i in range(10):
  print(i, end = ' ')
```

```python
# for 루프는 시퀸스(리스트, 튜플, 문자열 등)나 반복 가능한 객체(iterable)의 요소를 순회

for i in range(5):
  print(i, end = ' ')
print('\n')

count = 0
while count < 5:
  print(count, end = ' ')
  count += 1
```

```python
# 리스트를 순회하는 예

for i in [1,2,3,4,5]:
  print(i, end = ' ')
print('\n')

# 문자열을 순회하는 예
for i in "hello":
  print(i, end = '')
```

```python
# while 루프는 주어진 조건이 True 인 동안 코드 블록이 반복된다.
# 조건이 False이 되면 푸르가 종료

count = 0
sum = 0
while count < 5:
  sum += count
  print(sum, end = ' ')
  count += 1
```

```python
# Q. d에서 반복문을 사용하여 97 98 99을 출력하세요.

d = {'a':97, 'b':98, 'c':99}

for i in d:
  print(d[i], end = ' ')
```
