### Task3_0426. 주어진 숫자 리스트에서 특정 값보다 큰 요소가 하나라도 존재하는지 검사하고, 그 결과를 불 값으로 반환하는 함수를 작성하세요.

- numbers = [1, 2, 3, 10, 20]
- threshold = 15

```python
numbers = [1, 2, 3, 10, 20]

n_value = numbers[0:]

for i in n_value:
  if i > 15:
    print(i in numbers)
```

### Task4_0426. 주어진 문자열이 특정 문자열을 포함하는지 확인하고 결과를 불 값으로 반환하는 함수를 작성하세요.

- text = "hello world"
- substring = "world"

```python
text = "hello world"
substring = "world"

sp1 = set(text.split())
sp2 = set(substring.split())

inter1 = sp1 - sp2

if inter1 != sp1:
  print(f"{sp1}은 {sp2}를 포함합니다.")
else:
  print(f"{sp1}은 {sp2}를 포함하지 않습니다.")
```

### Task5_0426. 주어진 연도가 윤년인지 판별하는 함수를 작성하세요. 윤년은 다음의 조건을 만족해야 합니다:
- 4로 나누어 떨어진다.
- 100으로 나누어 떨어지지 않거나, 400으로 나누어 떨어지면 윤년이다.

year = 2020

True

```python
year = 2020

if year % 4 == 0 and year % 400 == 0 or year % 100 != 0:
    print(f'{year}년은 윤년입니다.', bool(1))
else:
    print(f'{year}년은 윤년이 아닙니다.', bool(0))
```

### Task6_0426. Calculator 클래스를 작성하고 4칙연산을 수행하는 객체 4개를 작성하여 결과를 출력하세요.

```python
# add, sub, mul, truediv, floordiv, mod

class Calc:
  def add(self,a,b,c,d):
    return a+b+c+d
  def sub(self,a,b,c,d):
    return a-b-c-d
  def mul(self,a,b,c,d):
    return a*b*c*d
  def floordiv(self,a,b,c,d):
    if b == 0 or c == 0 or d == 0:
      return 'Error : Get Out On My Face'
    else:
      return a/b/c/d

Obj1 = Calc()
print(Obj1.add(1,2,3,4))
Obj2 = Calc()
print(Obj2.sub(1,2,3,4))
Obj3 = Calc()
print(Obj3.floordiv(1,2,3,4))
Obj4 = Calc()
print(Obj4.mul(1,2,3,4))
```
