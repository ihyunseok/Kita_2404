#### Task3_0516. 주어진 문자열이 유효한 이메일 주소인지 확인
```
"example@example.com"
```

````python
user_text = input("이메일 주소를 기입 : ")

p = re.compile(r'.+(?=.)')
matches1 = re.findall(p, user_text)
print(matches1)
````

#### Task4_0516. 주어진 문자열이 올바른 전화번호 형식(010-0000-0000)인지 확인하세요.
```
"010-1234-5678"
```

````python
import re

Phonenum = input("전화번호를 입력해주세요. (010-1234-5678) : ")

p = re.compile(r'((\d+)[-](\d+)[-](\d+))')
m = p.search(Phonenum)

if len(m.group(2)) == 3 and len(m.group(3)) == 4 and len(m.group(4)) == 4 :
        print("올바른 전화번호 형식입니다.")
else:
        print("올바르지 않은 전화번호 형식입니다.")
````

#### Task7_0516. 주어진 문자열이 유효한 URL인지 확인하세요.
```
"https://www.example.com"
```

````python
import re

text = 'http://www.example.com'

p1 = re.compile(r'.+:')
m1 = p1.search(text)
mg1 = m1.group()

p2 = re.compile(r'(?=\.com).+')
m2 = p2.search(text)
mg2 = m2.group()

if mg1 == 'http:' and mg2 == '.com' :
      print(f"{text} : 유효한 URL 입니다.")
else :
      print(f"{text} : 유효하지 않는 URL 입니다. ")
````
