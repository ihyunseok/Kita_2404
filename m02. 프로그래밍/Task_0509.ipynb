#### Task1_0509. 주어진 텍스트에서 문자 빈도를 아래와 같이 출력하세요.(사용자 함수 사용)

가장 많이 나온 문자 5개를 뽑아 봤습니다.\
[e] -> [97]회 나타남\
[t] -> [85]회 나타남\
[a] -> [78]회 나타남\
[o] -> [73]회 나타남\
[s] -> [60]회 나타남

```python
def word_frequency(data):
    char_frequency = {}
    data = data.replace('\n', '').replace(' ', '')

    for char in data:
        if char in char_frequency:
            char_frequency[char] += 1
        else:
            char_frequency[char] = 1

    sorted_char_frequency = sorted(char_frequency.items(), key=lambda x: x[1], reverse=True)

    return sorted_char_frequency

char_frequency = word_frequency(data)
for char, count in char_frequency[:5]:
    print(f"[{char}] -> [{count}]회 나타남")
```

#### Task2_0509.  람다함수와 적합한 내장함수를 사용하여 아래와 같이 출력하세요.

[1,2,3,4,5] -> [1,4,9,16,25]\
[1,2,3,4,5] -> [1,2]

```python
# [1, 4, 9, 16, 25]

list_num = [1, 2, 3, 4, 5]
list1 = []

list2 = [lambda x: x*x for x in list_num]

for i in range(len(list_num)):
    list1.append(list2[i](list_num[i]))

print(list1)

# [1,2]

list_num = [1, 2, 3, 4, 5]

list3 = list(filter(lambda x: x <= 2, list_num))

print(list3)
```
